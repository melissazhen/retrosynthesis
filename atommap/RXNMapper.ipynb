```python
import pandas as pd
import random 
data = pd.read_csv("specialty chemicals.csv",encoding="gbk")
prod_smiles = []
rcts_smiles = []
rxn_smiles = []

reaction_data = list(set(list(data['Reaction'])))
random.seed(15)
random.shuffle(reaction_data)

for rxn_smi in reaction_data:
    try:
        split_smi = rxn_smi.split('>>')
    except:
        continue
    if rxn_smi[:2] == '>>':
        continue
    prod_smi = split_smi[-1]
    prod_smiles.append(prod_smi)
    rcts_smi = split_smi[0]
    rcts_smiles.append(rcts_smi)
    rxn_smiles.append(rxn_smi)

dataframe_explosive = pd.DataFrame({'Product':prod_smiles[:],'Reactant':rcts_smiles[:],'Rxn':rxn_smiles[:]})
dataframe_explosive.to_csv('specialty chemicals_train.csv',index=False)
```

```python
from rxnmapper import RXNMapper
import pandas as pd
rxn_mapper = RXNMapper()
df=pd.read_csv('specialty chemicalss_train.csv',encoding="gbk")
rxns = list(df['Rxn'][:])
products = list(df['Product'][:])
reactants = list(df['Reactant'][:])
results = []
results_id = []
wrong_id = []
wrong_rxn = []
count = 2
for i in range(len(rxns)):
    rxn = rxns[i]
    try:
        result = rxn_mapper.get_attention_guided_atom_maps([rxn])
        results.append(result[0])
        results_id.append(i)
    except:
        print(count)
        print(rxn)
        wrong_id.append(count)
        wrong_rxn.append(rxn)
    count += 1
```

```python
con = []
rxn = []
for item in results:
    con.append(item['confidence'])
    rxn.append(item['mapped_rxn'])
    

outframe = pd.DataFrame({'Confidence':con,'Mapped_rxn':rxn, 'Product':[products[i] for i in results_id],'Reactant':[reactants[i] for i in results_id], \
    'Rxn':[rxns[i] for i in results_id]})
outframe.to_csv('outframe.csv',index=False)
wrong_frame = pd.DataFrame({'wrong_id':wrong_id,'Wrong_rxn':wrong_rxn})
wrong_frame.to_csv('wrongframe.csv',index=False)
```
